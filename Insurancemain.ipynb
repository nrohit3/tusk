```python
m=insdata.shape[0]
n=insdata.shape[1]
ddof=(1)*(m-1)
alpha=0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(observed,expected)])
chi_square_statistic=chi_square[0]+chi_square[1]
p_value=1-chi2.cdf(x=chi_square_statistic,df=100)
```

```python
print("chi-square statistic:-",chi_square_statistic)

print('Significance level: ',alpha)

print('Degree of Freedom: ',ddof)

print('p-value:',p_value)
```

```python
if p_value<=alpha:

    print("Reject Null Hypothesis")

else:

    print("Accept Null Hypthesis")
```

```python
# 
l1=[]
l2=[]
for i in insdata:
    df=pd.crosstab(insdata[i],insdata['CARAVAN'])
    m=df.shape[0]-1
    n=df.shape[1]-1
    ddof=m*n
    observed=df.values
    val2=stats.chi2_contingency(df)
    expected=val2[3]
    from scipy.stats import chi2
    chi_square=sum([(o-e)**2./e for o,e in zip(observed,expected)])
    chi_square_statistic=chi_square[0]+chi_square[1]
    p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
    if p_value<=alpha:
        l1.append(i)
        l2.append(p_value)
        l3=pd.DataFrame(data=[l1,l2]).transpose()
```
